## 🧱 Step 1: Dataset Creation

### 📌 Overview

In this step, a dataset is manually created using a Python dictionary. This approach helps in understanding how structured data is formed before working with real-world datasets.

### 🎯 Objective

* Define features (columns) and records (rows)
* Convert raw data into a structured format

### 🛠️ Method

A dictionary is used where:

* Keys represent column names
* Values represent the data entries

This dictionary is then converted into a Pandas DataFrame.

### 📊 Features Included

* Name
* Age
* Gender
* Score
* Passed

Each row represents an individual record.

### 🔢 Custom Index

Custom index labels (`ID_1` to `ID_15`) are assigned to improve readability and demonstrate indexing techniques.


```python
import pandas as pd

data = {
    "Name": ["Abel", "Sara", "Kebede", "Marta", "Segni", "Alem", "Ruth", "Bonsa", "Liya", "Bedaso",
             "Hana", "Samuel", "Bethy", "Daniel", "Rahel"],
    "Age": [23, 21, 35, 28, 40, 20, 22, 30, 27, 33, 25, 29, 31, 38, 24],
    "Gender": ["M", "F", "M", "F", "M", "F", "F", "M", "F", "M", "F", "M", "F", "M", "F"],
    "Score": [85, 90, 78, 88, None, 95, 87, None, 84, 89, None, 76, 83, 79, None],
    "Passed": ["Yes", "Yes", "No", "Yes", None, "Yes", "Yes", None, "Yes", "Yes", None, "No", "Yes", "No", None]
}

df = pd.DataFrame(data)

# Custom index
df.index = [f"ID_{i}" for i in range(1, 16)]

print(df)
```

## 🔍 Step 2: Data Exploration

### 📌 Overview

Before cleaning or analyzing the dataset, it is important to understand its structure and identify potential issues such as missing values.

### 🛠️ Techniques Used

* `.head()` → Displays the first few rows of the dataset
* `.info()` → Provides information about data types and missing values
* `.describe()` → Generates statistical summary of numerical columns

### 🎯 Purpose

* Understand dataset structure
* Identify missing or inconsistent data
* Get an overview of data distribution


```python
print("First 5 rows:\n", df.head())
print("\nInfo:\n")
print(df.info())
print("\nSummary:\n", df.describe())
```

```python
print("=== BEFORE CLEANING ===")
print(df)
print("\nMissing Values:")
print(df.isnull().sum())
```

## 🧹 Step 3: Data Cleaning

### 📌 Overview

Real-world datasets often contain missing or incomplete values. In this step, missing data is handled to ensure consistency and accuracy.

### ⚠️ Identified Issues

* Missing values in the **Score** column
* Missing values in the **Passed** column

### 🛠️ Cleaning Methods

* **Score** → Filled using the mean (average value)
* **Passed** → Filled using the mode (most frequent value)

### 🎯 Purpose

* Prevent errors during analysis
* Improve data quality
* Maintain dataset completeness


```python

# Fill missing Score with mean
df["Score"] = df["Score"].fillna(df["Score"].mean())
df.fillna({"Score":df["Score"].mean()}, inplace=True)

# Fill missing Passed with mode
df["Passed"] = df["Passed"].fillna(df["Passed"].mode()[0])
df.fillna({"Passed":df["Passed"].mode()[0]}, inplace=True)
```

```python
print("\n=== AFTER CLEANING ===")
print(df)
print("\nMissing Values:")
print(df.isnull().sum())
```

## ⚙️ Step 4: Feature Engineering

### 📌 Overview

Feature engineering involves creating new features from existing data to improve analysis and interpretation.

### 🆕 New Feature Created

**Performance**

### 📊 Logic

* "High" → Score ≥ 85
* "Low" → Score < 85

### 🎯 Purpose

* Simplify interpretation of scores
* Enable easier comparison between individuals
* Enhance analytical insights


```python
# Create performance category
df["Performance"] = df["Score"].apply(lambda x: "High" if x >= 85 else "Low")
df["Performance"]

```

## 📊 Step 5: Data Analysis

### 📌 Overview

In this step, the dataset is analyzed to uncover patterns and relationships between variables.

### 🛠️ Techniques Used

Grouping operations using Pandas `groupby()` function.

### 🔍 Analysis Performed

* Average score by gender
* Pass rate distribution by gender

### 🎯 Purpose

* Identify trends in performance
* Compare groups within the dataset
* Extract meaningful insights


```python
# Average score by gender
print("\nAverage Score by Gender:\n", df.groupby("Gender")["Score"].mean())

# Pass rate by gender
print("\nPass Rate by Gender:\n", df.groupby("Gender")["Passed"].value_counts(normalize=True))

print("\nCleaned Dataset:\n", df)
```
