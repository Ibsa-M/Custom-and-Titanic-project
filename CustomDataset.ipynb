## 🧱 Part 1: Custom Dataset Creation

### 📌 Overview

In this section, a dataset was created manually using a Python dictionary and converted into a Pandas DataFrame. This step helps build a strong understanding of how structured data is formed and managed.

---

### 🎯 Objective

* Create a dataset with:

  * 5 columns (features)
  * 15 rows (records)
  * Custom index values
* Understand how data is organized in tabular form

---

### 🛠️ Method

A Python dictionary was used where:

* Keys represent column names
* Values represent the data entries

The dictionary was then converted into a DataFrame using Pandas.

---

### 📊 Dataset Description

The dataset includes the following features:

* **Name** → Represents individual names
* **Age** → Age of each individual
* **Gender** → Male or Female
* **Score** → Performance score
* **Passed** → Indicates whether the individual passed (Yes/No)

Each row represents a unique record.

---

### 🔢 Custom Index

Custom index labels (e.g., `ID_1`, `ID_2`, ..., `ID_15`) were assigned to improve readability and demonstrate indexing techniques in Pandas.

---

### 💡 Key Learning Points

* How to create a dataset from scratch
* Understanding DataFrame structure
* Importance of indexing
* How data is organized for analysis

---

### 📌 Conclusion

This step provides a foundational understanding of how datasets are built and structured before performing any real-world data analysis.


```python
import pandas as pd

data = {
    "Name": ["Abel", "Sara", "Kebede", "Marta", "Segni", "Alem", "Ruth", "Bonsa", "Liya", "Bedaso",
             "Hana", "Samuel", "Bethy", "Daniel", "Rahel"],
    "Age": [23, 21, 35, 28, 40, 20, 22, 30, 27, 33, 25, 29, 31, 38, 24],
    "Gender": ["M", "F", "M", "F", "M", "F", "F", "M", "F", "M", "F", "M", "F", "M", "F"],
    "Score": [85, 90, 78, 88, 92, 95, 87, 80, 84, 89, 91, 76, 83, 79, 86],
    "Passed": ["Yes", "Yes", "No", "Yes", "Yes", "Yes", "Yes", "No", "Yes", "Yes", "Yes", "No", "Yes", "No", "Yes"]
}

df = pd.DataFrame(data)

# Custom index
df.index = [f"ID_{i}" for i in range(1, 16)]

print(df)
```
