### Step 3: Data Cleaning

Handling missing values and removing unnecessary columns:
- Filled missing Age with median to maintain data consistency
- Filled Embarked with mode to fill missing categorical values
- Dropped Cabin column due to excessive missing data
- Removed duplicated rows to ensure data accuracy


 ### Droping unnecessary data
 After checking if the data is empty or not and Reloading it
 Drop duplicate rows
df.drop_duplicates(inplace=True)
Drop duplicates based on specific columns
df.drop_duplicates(subset=[list_of_columns], keep="first", inplace=True)

```python
print("=== BEFORE CLEANING ===")
print(df.head())

print("\nMissing Values Before Cleaning:")
print(df.isnull().sum())
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df = df.drop(columns=["Cabin"])
df = df.drop_duplicates()
```

```python
# print("\n=== AFTER CLEANING ===")
# print(df.head())

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())
```

### Step 4: Data Analysis

Analyzing survival patterns using groupby operations:
- Survival rate by gender
- Survival rate by class
- Average age per class
- Survival rate by age group

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby("Pclass")["Age"].mean()
```

```python
df["AgeGroup"] = pd.cut(df["Age"],
                       bins=[0, 12, 18, 60, 100],
                       labels=["Child", "Teen", "Adult", "Senior"])

df.groupby("AgeGroup")["Survived"].mean()
```

### Remember
The above message isn’t an error—it’s a FutureWarning. Pandas is just letting we know that the default behavior of groupby with categorical data will change in a future version. Your code is still running fine, but Pandas wants you to be explicit. but it works fine and give us what expected or aimed for
